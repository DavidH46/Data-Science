```python
#Import modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#Read the CSV file into a DataFrame
df = pd.read_csv('2008_all_states.csv')

#Compute percentiles of dem_share
percentile = np.percentile(df['dem_share'], [25, 50, 75])
print(percentile)

#Show boxplot. IQR = 25th~75th percentile
#Outliers: 1.5 IQR above 75th or below 25th percentile
_ = sns.boxplot(x=df['east_west'], y=df['dem_share'])
_ = plt.xlabel('region')
_ = plt.ylabel('percent of vote for Obama')
plt.show()

#Extract FL rows from the data
dem_share = df[['dem_share', 'state']]
dem_share_FL = dem_share.loc[dem_share['state'] == 'FL']

#Compute variance
print(np.var(dem_share_FL))
#Compute standard deviation
print(np.std(dem_share_FL))

#Finding relationship between dem_share and total_votes columns
#Extract columns to explore their correlation
dem_share = df['dem_share']
total_votes = df['total_votes']

#Plot settings
_ = plt.plot(total_votes/1000, dem_share, marker='.', linestyle='none')
_ = plt.xlabel('total votes (thousands)')
_ = plt.ylabel('percent of vote for Obama')
plt.show()

#Compute covariance
covariance_matrix = np.cov(dem_share, total_votes)
covariance = covariance_matrix[0,1]
print('Covariance: ' + str(covariance))
#Compute Pearson Coefficient: insignificant correlation
pearsoncoef = np.corrcoef(dem_share, total_votes)
print('Pearson Coefficient: ' + str(pearsoncoef[0, 1]))
```
